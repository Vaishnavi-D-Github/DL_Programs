```python
import numpy as np
inputs = np.array([1.0, 2.0, 3.0])
weights = np.array([1.0, -0.5, -0.5])
target = 1
bias = 0.0
learning_rate = 0.1
z = np.dot(inputs, weights) + bias
output = 1 if z >= 0 else 0
error = target - output
weights += learning_rate * error * inputs
bias += learning_rate * error
print("Updated Weights: ", weights)
print("Updated Bias: ", bias)
```

```python
z = float(input("Enter value of z: "))
def threshold (z):
  return 1 if z >= 0 else 0
def sigmoid (z):
  return 1/(1 + np.exp(-z))
def tanh(z):
  return np.tanh(z)
def relu(z):
  return max(0, z)

print("Threshold: ", threshold(z))
print("Sigmoid: ", sigmoid(z))
print("ReLu: ", relu(z))
print("Tanh", tanh(z))
```

```python
z = np.linspace(-10, 10, 100)
plt.plot(z, [threshold(i) for i in z], label='Threshold')
plt.plot(z, sigmoid(z), label='Sigmoid')
plt.plot(z, tanh(z), label='Tanh')
plt.plot(z, [relu(i) for i in z], label='ReLu')
plt.legend()
plt.show()
```
